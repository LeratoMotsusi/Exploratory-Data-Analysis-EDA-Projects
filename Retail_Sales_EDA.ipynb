```python
#Counts the number of occurances of each unique value in the Review Rating column.
display(Shopping_trends['Review Rating'].value_counts())
```

```python
#Creating Review Bucket Categories based on the review rating.
#Define the bins and labels for the review rating categories
bins = [1, 2, 3, 4, 5, float('inf')]
labels = ['Dissatisfied', 'Needs Improvement', 'Satisfied', 'Very Satisfied', 'Excellent']

# Create the new column 'Review_Category' by classifying 'Review Rating'
Shopping_trends['Review_Category'] = pd.cut(Shopping_trends['Review Rating'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Shopping_trends[['Review Rating', 'Review_Category']].head())
```

```python
#Identifies the number of unique values in the Previous Purchases column.
Shopping_trends['Previous Purchases'].nunique()
```

```python
#Creating Spender Categories based on the previous purchases.
#Define the bins and labels for the previous purchases categories
bins = [1, 2, 11, 31, 51]
labels = ['One-Time Buyer (1)', 'Repeat Buyer (2-10)', 'Regular Shopper (11-30)', 'Loyal Customer (31-50)']

# Create the new column 'Spender_Category' by classifying 'Previous Purchases'
Shopping_trends['Spender_Category'] = pd.cut(Shopping_trends['Previous Purchases'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Shopping_trends[['Previous Purchases', 'Spender_Category']].head())
```

```python
display(Shopping_trends)
```

```python
#Creating Spender Categories based on the previous purchases.
#Define the bins and labels for the previous purchases categories
bins = [1, 31, 61, 100]
labels = ['Low Spender ($30)', 'Medium Spender ($60)', 'High Spender($100)']

# Create the new column 'Purchase_Category' by classifying 'Purchase Amount (USD)'
Shopping_trends['Purchase_Category'] = pd.cut(Shopping_trends['Purchase Amount (USD)'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Shopping_trends[['Purchase Amount (USD)', 'Purchase_Category']].head())
```

```python
display(Shopping_trends)
```

```python
# Convert the 'processdate' column to datetime objects
Shopping_trends['processdate'] = pd.to_datetime(Shopping_trends['processdate'])

# Extract the month name and create a new column 'Month_Name'
Shopping_trends['Month_Name'] = Shopping_trends['processdate'].dt.month_name()

# Display the first few rows with the new column
display(Shopping_trends[['processdate', 'Month_Name']].head())
```

```python
# Extract the year and month and format them as 'YYYY-MM'
Shopping_trends['Month ID'] = Shopping_trends['processdate'].dt.strftime('%Y-%m')

# Display the first few rows with the new column
display(Shopping_trends[['processdate', 'Month ID']].head())
```

```python
display(Shopping_trends)
```

```python
# Extract the time and create a new column 'Process_Time'
Shopping_trends['Process_Time'] = Shopping_trends['processdate'].dt.time

# Display the first few rows with the new column
display(Shopping_trends[['processdate', 'Process_Time']].head())
```

```python
#Creating Time Categories based on the Process Time.
#Define the bins and labels for the previous purchases categories
import datetime
bins = [datetime.time(0, 0, 0), datetime.time(6, 0, 0), datetime.time(12, 0, 0), datetime.time(18, 0, 0), datetime.time(23, 59, 59)]
labels = ['Early Morning(Midnight-6am)', 'Morning(6am-12noon)', 'Afternoon(12noon - 6pm)', 'Evening(6pm-Midnight)']

# Create the new column 'Process Time Category' by classifying 'Process_Time'
Shopping_trends['Process_Time_Category'] = pd.cut(Shopping_trends['Process_Time'], bins=bins, labels=labels, right=False, ordered=False)

# Display the first few rows with the new column
display(Shopping_trends[['Process_Time', 'Process_Time_Category']].head())
```

```python
display(Shopping_trends)
```

```python
#Creating Age Categories based on the Age column.
#Define the bins and labels for the previous purchases categories
bins = [18, 31, 46, 61, 70]
labels = ['Young Adult (18-30)', 'Middle Age (31-45)', 'Matured (46-60)', 'Pensioner (61-70)']

# Create the new column 'Age_Category' by classifying 'Age'
Shopping_trends['Age_Category'] = pd.cut(Shopping_trends['Age'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Shopping_trends[['Age', 'Age_Category']].head())
```

```python
display(Shopping_trends)
```

```python
# Extract the day of the week name and create a new column 'Day_of_Week'
Shopping_trends['Day_of_Week'] = Shopping_trends['processdate'].dt.day_name()

# Display the first few rows with the new column
display(Shopping_trends[['processdate', 'Day_of_Week']].head())
```

```python
# Create a new column 'Day_Type' based on 'Day_of_Week'
# Define weekdays and weekends
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday']
weekends = ['Friday', 'Saturday', 'Sunday']

# Use .loc to assign 'Weekday' or 'Weekend' based on the 'Day_of_Week' column
Shopping_trends.loc[Shopping_trends['Day_of_Week'].isin(weekdays), 'Day_Type'] = 'Weekday'
Shopping_trends.loc[Shopping_trends['Day_of_Week'].isin(weekends), 'Day_Type'] = 'Weekend'

# Display the first few rows with the new column
display(Shopping_trends[['Day_of_Week', 'Day_Type']].head())
```
