```python
#Creating Age Categories based on the Age column.
#Define the bins and labels for the previous purchases categories
bins = [17, 18, 31, 46, 61, 70]
labels = ['Youth (0-18)', 'Young Adult (18-30)', 'Middle Age (31-45)', 'Matured (46-60)', 'Pensioner (61-70)']

# Create the new column 'Age_Category' by classifying 'Age'
Retail_sales['Age_Category'] = pd.cut(Retail_sales['Age'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Retail_sales[['Age', 'Age_Category']].head())
```

```python
#Creating Revenue Categories based on the Revenue column.
#Define the bins and labels for the previous purchases categories
bins = [0, 400, 800, 1200, 1600, 2000] # Added 0 to the bins
labels = ['Budget Spender(0-R400)', 'Low Spender(R401-R800)', 'Mid Spender(R801-R1200)', 'High Spender(R1201-R1600)', 'Premium Spender(R1601-R2000)']

# Create the new column 'Revenue_Category' by classifying 'Revenue'
Retail_sales['Revenue_Category'] = pd.cut(Retail_sales['Revenue'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(Retail_sales[['Revenue', 'Revenue_Category']].head())
```

```python
# Find the lowest value in the 'Revenue' column
lowest_revenue = Retail_sales['Revenue'].min()

# Find the highest value in the 'Revenue' column
highest_revenue = Retail_sales['Revenue'].max()

# Display the results
print(f"Lowest Revenue: {lowest_revenue}")
print(f"Highest Revenue: {highest_revenue}")
```
